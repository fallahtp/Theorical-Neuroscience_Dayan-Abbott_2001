```python
# --- Imports ---
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- Setting parameters ---
A = 1
dispatiry = np.arange(-np.pi, np.pi, 0.1)
```

```python
# --- Calculating Complex cell response ---
L = (A**2) * (np.cos(dispatiry/2)**2)
# Plot
plt.plot(dispatiry, L)
plt.xlabel('\u03A6'+'L - ' + '\u03A6'+'R')
plt.ylabel('L1^2 + L2^2')
plt.show()

```
